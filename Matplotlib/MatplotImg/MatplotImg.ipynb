```python
%matplotlib inline
import matplotlib.pyplot as plt 
import matplotlib.image as mpimg
```

```python
#Her R,G,B değerleri için 8 bit değeri oluşturulur ve result değeleri 8 bit için 0.0 ile 1.0 arasında ölçeklendirilir.
img = mpimg.imread("togg.jpg")
print(img)
```

```python
#Dizide elde ettiğimiz sayısal verilerden geri dönerek tekrar resim oluşturna işlemini imshow ile gerçekleştirebilirz. 
imgplot = plt.imshow(img)
```

```python
t_img=img[:,:,0]
plt.imshow(img)
plt.show()
```

```python
plt.imshow(t_img,cmap="hot")
```

```python
#set_cmap ile çizim nesnelerindeki renk haritalarını değiştirme
# bu iki komutun aynı hücrede olması gerekliliği durumu mevcuttur.
imgplot = plt.imshow(t_img)
imgplot.set_cmap("nipy_spectral")
```

```python
#Renk skalası bir rengin hangi değeri ifade ettiğini gösteren bar çubuk
imgplot=plt.imshow(t_img)
plt.colorbar()
```

```python
#Interpolation  farklı matematiksel şemalara göre bir pikselin renginin veya değerinin olmasını gerektiğini hesaplar.
#Görüntü grafiği nesnesini set_clim() metodunu çağırarak da yapabilirsiniz.
fig= plt.figure()
x = fig.add_subplot(1,2,1)
imgplot= plt.imshow(t_img)
x.set_title="Öncesi"
plt.colorbar(ticks=[0.1, 0.3, 0.5, 0.7], orientation='horizontal')
fig= plt.figure()
x = fig.add_subplot(1,2,2)
imgplot= plt.imshow(t_img)
imgplot.set_clim(0.0, 0.7)
x.set_title="Sonrası"
plt.colorbar(ticks=[0.1, 0.3, 0.5, 0.7], orientation='horizontal')
```
