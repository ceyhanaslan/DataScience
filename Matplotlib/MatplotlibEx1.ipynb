```python
#Kolonları kullanabilmek için boşlukları silerek _ kullandık
#Sonrasında DataFrame 'i rankings_pd_2019 datasına atadık.

data2019.columns = data2019.columns.str.replace(' ', '_')
rankings_pd_2019 = pd.DataFrame(data2019)
rankings_pd_2019.columns
```

```python
rankings_pd_2019.head()
```

```python
#Line Plot 
#Eğer x axis zaman ise Line plot kullanmak mantıklıdır
rankings_pd_2019.Social_support.plot(color = 'purple',label = 'Social_support',linewidth=1, alpha = 0.8,grid = True,linestyle = '-.')
rankings_pd_2019.Gdp_per_capita.plot(kind = 'line', color = 'blue',label = 'Gdp_per_capita',linewidth=1,alpha = 0.5,grid = True,linestyle = ':')
plt.legend(loc="upper right")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("World Happiness Line Plot")
plt.show()
```

```python
#Scatter plot iki tane varible arasında korelasyon var ise kullanmak iyidir
rankings_pd_2019.plot(kind="scatter",x="Social_support",y="Gdp_per_capita",alpha=0.5,color="blue")
plt.xlabel("Social_support")
plt.ylabel("Gdp_per_capita")
plt.show()

#Alttaki sorgu ile üstteki arasında fark yoktur iki kullanımda scatter için aynıdır.
#plt.scatter(rankings_pd_2019.Social_support,rankings_pd_2019.Gdp_per_capita,alpha=0.5,color="blue")
#plt.show()
```
